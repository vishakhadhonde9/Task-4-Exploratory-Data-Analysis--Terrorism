#Number of Attack per Year

```python
year_data = dataset['iyear'].value_counts().sort_values()
year_data
```

```python
year = dataset['iyear'].unique()
years_count = dataset['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
pd.crosstab(dataset.iyear, dataset.region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(dataset['country'].value_counts()[:10].index,dataset['country'].value_counts()[:10].values,palette='YlOrBr_r')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

#Type of Attack

```python
dataset['attacktype1_txt'].value_counts().plot(kind='bar',figsize=(20,10),color='magenta')
plt.xticks(rotation = 50)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Name of attacktype",fontsize=20)
plt.show()
```

#Terrorist Group Involved in most Attacks

```python
gr_data=dataset['gname'].value_counts().nlargest(n=10)
gr_data
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(x = gr_data.values, y = gr_data.index)
plt.title('Most freq. Terrorist group involved in Attack')
plt.xlabel('Attacks count')
plt.ylabel('Groups')
plt.xticks(rotation = 90)
plt.show()
```

#Mostly affected Cities

```python
city_data = dataset['city'].value_counts().head(10)
city_data
```

```python
dataset['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='red')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)
plt.show()
```

#Mostly affected States

```python
states_data = dataset['provstate'].value_counts().head(10)
states_data
```

```python
dataset['provstate'].value_counts().to_frame().sort_values('provstate',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation = 50)
plt.xlabel("States",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected states",fontsize=20)
plt.show()
```

#Mostly affected Region

```python
reg_data =dataset['region'].value_counts().head(10)
reg_data
```

```python
dataset['region'].value_counts().to_frame().sort_values('region',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='purple')
plt.xticks(rotation = 50)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected Region",fontsize=20)
plt.show()
```

Number of Wounded

```python
dataset[['attacktype1_txt','nwound']].groupby(["attacktype1_txt"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

Died people in Attack

```python
df = dataset[['iyear','nkill']].groupby(['iyear']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title("People Died in Attack",fontsize=25)
plt.ylabel("Number of killed peope",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

Number of Kills

```python
dataset[['attacktype1_txt','nkill']].groupby(["attacktype1_txt"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkslateblue'])
plt.xticks(rotation=50)
plt.title("Number of killed ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```
